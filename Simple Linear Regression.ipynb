```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Advertising.csv')
```

```python
df.head()
```

```python
df['total_advertising'] = df['TV'] + df['radio'] + df['newspaper'] + df['sales']
```

```python
df.head(10)
```

```python
sns.scatterplot(data=df, x='total_advertising', y='sales')
```

```python
#there is clearly a positive regression fit , the more we spend on advertising, the more sales convert
```

```python
# we can plot a regression line that plots a line of best fit for our viz
```

```python
sns.regplot(data=df, y='sales', x='total_advertising')
```

```python
X= df['total_advertising']
```

```python
y= df['sales']
```

```python
# y = mx + b

np.polyfit(X,y, deg=1)
```

```python
potential_spend = np.linspace(0, 500, 100)
```

```python
predicted_sales = 0.04710145 * potential_spend + 3.90119955
```

```python
sns.scatterplot(data= df, x='total_advertising', y='sales' )
plt.plot(potential_spend, predicted_sales, color='red')
```

```python
#suppose you have a total_advertising spend of 300, use eq and visual to proove the sales you'll get

spend = 300
predicted_sales = 0.04710145 * spend + 3.90119955
```

```python
predicted_sales
```

```python
# let us use another model, where things are not linear, but polynomial
```

```python
# The linear fit -> y= B1x + B0
# The polynomial fit -> y= B3x**3 + B2x**2 + B1x + B0

np.polyfit(X,y,3)
```

```python
pot_spend = np.linspace(0,500,100)
```

```python
pred_sales = 2.64617556e-07* pot_spend **3 + -1.67427728e-04 * pot_spend ** 2 + 7.71590458e-02 * pot_spend + 2.50495056e+00
```

```python
sns.scatterplot(data=df, x='total_advertising', y='sales')
```

```python
sns.scatterplot(data=df, x='total_advertising', y='sales')
plt.plot(pot_spend, pred_sales, color='red')
```
